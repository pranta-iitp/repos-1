```python
import pandas as pd
from sklearn.model_selection import train_test_split
# Step 1: Load the dataset
df = pd.read_csv("titanic_toy.csv")
# Step 2: Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["Survived"]), df["Survived"], test_size=0.2)
# Step 3: Calculate mean and median for Age and Fare columns from the training data
mean_age = X_train["Age"].mean()
median_age = X_train["Age"].median()
mean_fare = X_train["Fare"].mean()
median_fare = X_train["Fare"].median()
# Step 4: Perform mean and median imputation for the training data
X_train["Age_median"] = X_train["Age"].fillna(median_age)  # Median Imputation for Age
X_train["Age_mean"] = X_train["Age"].fillna(mean_age)      # Mean Imputation for Age
X_train["Fare_median"] = X_train["Fare"].fillna(median_fare)  # Median Imputation for Fare
X_train["Fare_mean"] = X_train["Fare"].fillna(mean_fare)      # Mean Imputation for Fare
# Step 5: Perform mean and median imputation for the test data using training statistics
X_test["Age_median"] = X_test["Age"].fillna(median_age)  # Median Imputation for Age
X_test["Age_mean"] = X_test["Age"].fillna(mean_age)      # Mean Imputation for Age
X_test["Fare_median"] = X_test["Fare"].fillna(median_fare)  # Median Imputation for Fare
X_test["Fare_mean"] = X_test["Fare"].fillna(mean_fare)      # Mean Imputation for Fare
```
