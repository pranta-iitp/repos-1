```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=["python","c","c++","java"]
y=[70,80,85,65]
z=[40,50,44,18]
p = np.arange(len(x))
p1 = [i+0.2 for i in p]

plt.xlabel("langauge",fontsize=15)
plt.ylabel("No.",fontsize=15)

plt.bar(p,y,color="r",width=0.2)
plt.bar(p1,z,color="g",width=0.2)
plt.xticks(p+0.2/2,x)
plt.show()
```
