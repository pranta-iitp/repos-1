```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
ohe.fit(X_train[['fuel','owner']])
X_train_new = ohe.transform(X_train[['fuel','owner']])
X_test_new = ohe.transform(X_test[['fuel','owner']])
# Ensure X_train and X_test are pandas DataFrames
X_train_subset = X_train[["brand", "km_driven"]].values  # Extract values for other columns
X_test_subset = X_test[["brand", "km_driven"]].values
# Concatenate horizontally using np.hstack
X_train_combined = np.hstack((X_train_subset, X_train_new))
X_test_combined = np.hstack((X_test_subset, X_test_new))
X_train = pd.DataFrame(X_train_combined) # converting to dataframe
X_test = pd.DataFrame(X_test_combined)
```
