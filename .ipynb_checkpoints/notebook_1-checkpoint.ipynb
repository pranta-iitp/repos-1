```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example data: 'age' and 'salary'
data = {
    'age': [22, 25, 29, 32, 40, 45, 50, 23, 27, 33, 38, 47, 55],
    'salary': [20000, 25000, 30000, 35000, 40000, 45000, 50000, 22000, 27000, 34000, 38000, 46000, 55000]
}

# Plot age distribution with salary distribution overlay
sns.histplot(data['age'], kde=True, label='Age', color='blue')
sns.histplot(data['salary'], kde=True, label='Salary', color='orange')

plt.legend()
plt.show()
```
