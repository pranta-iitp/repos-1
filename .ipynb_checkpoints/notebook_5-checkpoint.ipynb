```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
# Load dataset
df = pd.read_csv("cars.csv")
# Split data
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:4], df.iloc[:, -1], test_size=0.2, random_state=2)
# Apply ColumnTransformer with OneHotEncoder
transformer = ColumnTransformer(transformers=[
    ('tnf3', OneHotEncoder(sparse_output=False, drop='first', dtype=np.int32), ['brand','fuel', 'owner'])
], remainder='passthrough')
# Fit and transform training data
X_train_new = transformer.fit_transform(X_train)
# Generate column names for the new DataFrame
ohe_features = transformer.named_transformers_['tnf3'].get_feature_names_out(['brand','fuel', 'owner'])
passthrough_features = X_train.columns.difference(['brand','fuel', 'owner'], sort=False)
all_features = np.hstack((ohe_features, passthrough_features))
# Create DataFrame with proper column names
X_train_new = pd.DataFrame(X_train_new, columns=all_features)
# Display the transformed DataFrame
print(X_train_new.head())
```
