```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day=["Monday","Tuesday","Wednesday"]
w1=[5000,5600,5900]
w2=[3000,4500,6000]
w3=[4500,2500,4800]
width=0.2
p=np.arange(len(day))
p1=[i+width for i in p]
p2=[i+width for i in p1]
plt.bar(p,w1,color="r",width=0.2,label="week1")
plt.bar(p1,w2,color="g",width=0.2,label="week2")
plt.bar(p2,w3,color="y",width=0.2,label="week3")
plt.xticks(p+width,day,rotation=90)
plt.legend()
plt.show()
```
